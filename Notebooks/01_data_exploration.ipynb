```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv("../Data/NASA_Exoplanet_Archive_Data.csv")
data.head()
```

```python
# Informacion de estadisticas descriptivas
data.describe()
```

```python
# Informacion general del dataframe
data.info()
```

```python
# ordenar desc por cantidad de valores nulos
df_sorted = data.isnull().sum().sort_values(ascending=False)
print(df_sorted.head(40))
```

```python
# Mostrar los valores de la columna discoverymethod

print(data.discoverymethod.value_counts())
```

```python
# Mostrar los valores de la columna pl_eqterr1
print(data.pl_eqterr1.value_counts())
```

```python
# graficar la cantidad de planetas descubiertos por cada metodo de descubrimiento
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x="pl_orbper", order=data["pl_orbper"].value_counts().index)
plt.xticks(rotation=90)
plt.title("Cantidad de planetas descubiertos por cada método de descubrimiento")
plt.xlabel("Masa del planeta")
plt.ylabel("Cantidad de planetas")
```
